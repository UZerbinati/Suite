# Home Work 1: "Accuracy and Stability (Automatic Differentiation)"

In this notebook we study the behaviour of finite difference approximation of the derivative of a real function, $f(x): \mathbb{R}\to \mathbb{R}$. In particular we will study forward finite difference and centered finite difference, respectively defined by the equations:
$$f'(x) \approx \frac{f(x+h)-f(x)}{h}$$
$$f'(x) \approx \frac{f(x+h)-f(x-h)}{2h}$$
for a small $h\in \mathbb{R}_{>0}$, in particular we will show that in a given range choosing a smaller $h$ improves the approximations. For the previous approximations we have the following estimates for the discretization error:
$$\Big|\;f'(x)- \frac{f(x+h)-f(x)}{h}\;\Big|\approx \frac{h}{2} |f''(x)| \label{eq:forwardest}$$
$$\Big|\;f'(x)- \frac{f(x+h)-f(x-h)}{2h}\;\Big|\approx \frac{h^2}{6} |f'''(x)|$$
We begin studying the function $f(x) = sin(x)$ at $x=1.2$ and the evolution of the approximation error from $10^{-1}\dots 10^{-16}$ since $f'(x) = cos(x)$.

```python
#We set the path for importing the suite C++ library 
import sys
sys.path.append('../../../Py/Build')

import matplotlib.pyplot as plt
from math import cos, sin, pi ,log 
from suite import *

def f(x):
    return sin(x);
a = 0; b = 2*pi; N = 100; hx= (b-a)/(N-1)
X = [a+(n*hx) for n in range(N)]
Y = [f(x) for x in X]
plt.plot(X,Y)
plt.legend(["sin(x)"])
```

We now use the suite C++ library to compute the derivative of the function $f(x)$ and we then compare it to the exact derivative.

```python
df_h = Diff()
df_h.push(f)
Y2 = [df_h.call(x) for x in X]
plt.plot(X,Y2)
plt.legend(["cos(x)"])
```

```python
def df(x):
    return cos(x);

H = [];
E = [];
Est = [];
for n in range(1,8):
    steph = 10**(-n);
    H = H + [steph]
    df_h.set_h(steph);
    E = E + [abs(df_h.call(1.2)-df(1.2))]
    Est = Est + [0.5*abs(-sin(1.2))*steph]
slope = log(E[0]/E[-1])/log(H[0]/H[-1])
plt.loglog(H,E,"*-")
plt.loglog(H,Est,"--")
plt.legend(["Error of forward finite difference","Error estimate"])
```

In particular we can observe that the error estimate provided in $\eqref{eq:forwardest}$ overlap perfectly on the true error, in the range $10^{-1}$ to $10^{-7}$. In particular the error "line" has slope {{slope}}. Some thing different occurs if we enlarge the range of h, in particular in the next example we will focus our attention to a h as small as $10^{-17}$.

```python
H = [];
E = [];
Est = [];
for n in range(1,17):
    steph = 10**(-n);
    H = H + [steph]
    df_h.set_h(steph);
    E = E + [abs(df_h.call(1.2)-df(1.2))]
    Est = Est + [0.5*abs(-sin(1.2))*steph]
plt.loglog(H,E,"*-")
plt.loglog(H,Est,"--")
plt.legend(["Error of forward finite difference","Error estimate"])
```

We observe that for $h$ smaller then e-8 the approximation error stop diminishing and actually start getting bigger. This is due to the fact that the round-off error becomes predominant with respect to the discretization error. In particular we observe _cancellation_, i.e. since we are taking a very small $h$ we end up having $f(x+h)$ that differs to $f(x)$ only by roundoff error and this error is amplified by a big factor $\frac{1}{h}$.

We can do the same numerical experiment using centered finite difference, to observe that cancellation is a common problem among all the finite difference schemes. In particular we observe that the cancellation phenomena starts sooner in the case of centered finite difference, this is due to the fact that here the round off error is amplified by a factor $\frac{1}{h^2}$.

```python
H = [];
EC = [];
ECst = [];
Est = [];
df_h.set_type("centered")
for n in range(1,17):
    steph = 10**(-n);
    H = H + [steph]
    df_h.set_h(steph);
    EC = EC + [abs(df_h.call(1.2)-df(1.2))]
    Est = Est + [0.5*abs(-sin(1.2))*steph]
    ECst = ECst + [(1/6)*abs(-cos(1.2))*(steph**2)]
plt.loglog(H,EC,"*-")
plt.loglog(H,E,"*-")
plt.loglog(H,Est,"--")
plt.loglog(H,ECst,"--")
plt.legend(["Error of centered finite difference",
            "Error of forward finite difference",
            "Error estimate forward fintie difference",
            "Error estimate centered fintie difference"])
```

One might wonder if using the relative error instead of the absolute error yields a different result, to answer this particular question we compute the above numerical experiment using the relative error instead of the absolute error. In particular we observe that there is no particular difference between the absolute error and the relative error in the evaluation of this numerical scheme.

```python
H = [];
E = [];
Erel = [];
Est = [];
df_h.set_type("forward")
for n in range(1,17):
    steph = 10**(-n);
    H = H + [steph]
    df_h.set_h(steph);
    E = E + [abs(df_h.call(1.2)-df(1.2))]
    Erel = Erel + [abs(1- df_h.call(1.2)/df(1.2))]
    Est = Est + [0.5*abs(-sin(1.2))*steph]
plt.loglog(H,Erel,"*-")
plt.loglog(H,E,"*-")
plt.loglog(H,Est,"--")
plt.legend(["Error (relative) of forward finite difference",
            "Error of forward finite difference",
            "Error estimate"])
```

Last but not least we compute the above numerical experiment for $g(x)=386.98 cos(x)$ and $x=1.2$ to show the importance of the absolute value of the second derivative. 

```python
def g(x):
    return 386.98*cos(x);
def dg(x):
    return -386.98*sin(x);
def d2g(x):
    return -386.98*cos(x);
dg_h = Diff();
dg_h.push(g);

H = [];
E = [];
Erel = [];
Est = [];
dg_h.set_type("forward")
for n in range(1,17):
    steph = 10**(-n);
    H = H + [steph]
    dg_h.set_h(steph);
    E = E + [abs(dg_h.call(1.2)-dg(1.2))]
    Erel = Erel + [abs(1- dg_h.call(1.2)/dg(1.2))]
    Est = Est + [0.5*abs(d2g(1.2))*steph]
plt.loglog(H,Erel,"*-")
plt.loglog(H,E,"*-")
plt.loglog(H,Est,"--")
plt.legend(["Error (relative) of forward finite difference",
            "Error of forward finite difference",
            "Error estimate"])
```

We observe that when measuring the absolute error, we still observe the cancellation error that begins at $10^{-8}$ since using trigonometric function $f(x+h)-f(h)$ becomes dominated by the round off error at $10^{-8}$. But the error curve is shifted above since as we could see in equation $\eqref{eq:forwardest}$ that the error estimate is multiplied by a factor of $10^{2}$. To understand the behaviour of the relative error we do a comparison of the relative error for g and f.

```python
H = [];
Eg = [];
Egrel = [];
Ef = [];
Efrel = [];
Estf = []
Estg = []

relf = abs(0.5*((-sin(1.2))/df(1.2)))
relg = abs(0.5*(d2g(1.2)/dg(1.2)))

dg_h.set_type("forward")
for n in range(1,17):
    steph = 10**(-n);
    H = H + [steph]
    Estg = Estg + [relg*steph]
    Estf = Estf + [relf*steph]
    dg_h.set_h(steph);
    df_h.set_h(steph);
    Eg = Eg + [abs(dg_h.call(1.2)-dg(1.2))]
    Ef = Ef + [abs(df_h.call(1.2)-df(1.2))]
    Egrel = Egrel + [abs(1- dg_h.call(1.2)/dg(1.2))]
    Efrel = Efrel + [abs(1- df_h.call(1.2)/df(1.2))]


plt.loglog(H,Egrel,"*-")
plt.loglog(H,Estg,"--")
plt.loglog(H,Efrel,"*-")
plt.loglog(H,Estf,"--")
plt.legend(["Relative error estimate for g'(x)",
            "Relative error estimate for g'(x)",
            "Relative error estimate for f'(x)",
            "Relative error estimate for f'(x)"])
```

We observe that the shift of relative error is less affected by the "magnitude" of the second derivative, this is due to fact that with some manipulation we can get the following estimate for the relative error:
$$e_{rel}(x) = \Big | 1 - \frac{f(x+h)-f(x)}{h}\frac{1}{f'(x)} \Big| \approx \Bigg| 1 - \Big(f'(x)+\frac{1}{2}f''(x)h \Big)\frac{1}{f'(x)} \Bigg|\approx \Big| \frac{1}{2} \frac{f''(x)}{f'(x)} h\Big|$$
there fore the important quantity for the relative error estimate is $\Big| \frac{1}{2} \frac{f''(x)}{f'(x)} \Big|$ that computed for $f$ is {{relf}} and for $g$ is {{relg}}.

## Appendix - Suite Diff Class

```python
#ifndef DIFFHEADERDEF
#define DIFFHEADERDEF

#include <iostream>
#include <string>
#include <functional>

using namespace std;

class Diff
{
	private:
		double h;
		string type;
		std::function <double(double)> f;
	public:

		Diff();
		double call(double x);
		void push(std::function <double(double)> g);
		void set_h(double step);
		void set_type(string fd);
};

#endif
```

```python
#include <cmath>
#include "diff.hpp"

double dummy(double x){
	return 0;
}
Diff::Diff(){
	f = dummy;
	h = 0.001;
	type="forward";
}

double Diff::call(double x){
	double value;
	if(type ==  "forward"){
		value= (1/h)*(f(x+h)-f(x));
	}
	if(type == "backward"){
		value=(1/h)*(f(x)-f(x-h));
	}
	if(type == "centered"){
		value=(1/(2*h))*(f(x+h)-f(x-h)); 
	}
	return value;
}
void Diff::push(std::function<double(double)> g){
	f = g;	
}
void Diff::set_h(double step){
	h = step;
}
void Diff::set_type(string fd){
	type = fd;
}
```
