```python
Geo = Geometry(2);
def distSquare(P):
    """
    y2 -> *----------*
           |          |
           |          |
           |          |
    y1 -> *----------*
           ^          ^
           |x1        |x2
    """
    x1 = 0; x2 = 1; y1 = 0; y2=1;
    return max([P[0]-x2,x1-P[0],P[1]-y2,y1-P[1]])
Geo.add(distSquare)
GeoDraw(Geo,[[-1,1],[-1,1]],20)
```

```python
Geo = Geometry(2);
def distSquare(P):
    """
    y2 -> *----------*
           |          |
           |          |
           |          |
    y1 -> *----------*
           ^          ^
           |x1        |x2
    """
    x1 = 0; x2 = 1; y1 = 0; y2=1;
    return max([P[0]-x2,x1-P[0],P[1]-y2,y1-P[1]])
def distSquare2(P):
    """
    y2 -> *----------*
           |          |
           |          |
           |          |
    y1 -> *----------*
           ^          ^
           |x1        |x2
    """
    x1 = -1; x2 = 0; y1 = -1; y2=0;
    return max([P[0]-x2,x1-P[0],P[1]-y2,y1-P[1]])
Geo.add(distSquare)
Geo.inter(distSquare2)
GeoDraw(Geo,[[-1,1],[-1,1]],20)
```

```python
Geo = Geometry(2);
def distSquare(P):
    """
    y2 -> *----------*
           |          |
           |          |
           |          |
    y1 -> *----------*
           ^          ^
           |x1        |x2
    """
    x1 = 0; x2 = 1; y1 = 0; y2=1;
    return max([P[0]-x2,x1-P[0],P[1]-y2,y1-P[1]])
def distCirc(P):
    return (P[0]**2+P[1]**2)-0.25;
Geo.add(distSquare)
Geo.sub(distCirc)
GeoDraw(Geo,[[-1,1],[-1,1]],20)
```
